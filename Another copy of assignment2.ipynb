```python
import numpy as np
def matrix_multiply(A, B):
    return np.matmul(A, B)
A = np.array([[1, 3],
              [5, 7]])
B = np.array([[2, 6],
              [4, 8]])
result = matrix_multiply(A, B)
print("Matrix multiplication result:\n", result)
```
